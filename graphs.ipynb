```python
import csv
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
epoch = []
acc = []
loss = []
```

```python
with open('models/emotion/training.log') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 3
    for row in csv_reader:
        line_count +=1
        if line_count % 12 == 0 and line_count !=6:
            acc.append(round(float(row[1]),2))
            loss.append(round(float(row[2]),2))
```

```python
for i in range(0,132,6):
    epoch.append(np.float64(0.0) + i)            
f1 = plt.figure()
f1.set_size_inches(10, 6)
f1.canvas.set_window_title('Training Log')
gs1 = gridspec.GridSpec(1, 2)
a1 = plt.subplot(gs1[0])
a1.set_title("Training Accuracy")
a1.plot(epoch, acc)
a1.set(xlabel='Epoch', ylabel='Accuracy')
```

```python
a2 = plt.subplot(gs1[1])
a2.set_title("Training Loss")
a2.plot(epoch, loss)
a2.set(xlabel='Epoch', ylabel='Loss')
plt.show()
```
