# Import Packages

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

#  Load & Display Image

```python
img = cv.imread('test_images/group_people.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

# Load Classifier

```python
face_cascade = cv.CascadeClassifier('../models/detection/haarcascade_frontalface_default.xml')
```

# Detect Faces

```python
faces = face_cascade.detectMultiScale(img_gray, 1.3, 5)
print('Faces Found: ', len(faces))
```

# Draw Detections

```python
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = img_gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
```

# Display & Save Result

```python
plt.imshow(img)
cv.imwrite('detection_results/latest_image.png', img)
```
