```python
#import opencv
import cv2 as cv
#import matplotlib to implement plotting functionality
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading in an image
img = cv.imread('test_images/group_people.png')
```

```python
#display image
plt.imshow(img)
```

```python
#write image to disk
cv.imwrite('BGROpenCV.png', img)
```

```python
#convert BGR to RGB to remove blue tinge
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#convert BGR to GRAY
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
#resizing image
larger_image = cv.resize(img_rgb,(1000,1000))
plt.imshow(larger_image)
```

```python
#rotate image
rows, cols = img_rgb.shape[:2]
center = cv.getRotationMatrix2D((cols/2,rows/2),180,1)
img_rotated = cv.warpAffine(img_rgb,center,(cols,rows))
plt.imshow(img_rotated)
```

```python
#edge detection
edges = cv.Canny(img_rgb,100,200)
plt.imshow(edges)
```
